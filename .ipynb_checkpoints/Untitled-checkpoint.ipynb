```python
import pandas as pd
```

```python
Trades = pd.read_csv("Crypto.csv" , index_col=0)
```

```python
Trades.rename(columns = {'عدد الاهداف المحققة':'Ac'}, inplace = True) 
```

```python
def calc_prof(en , tps , ac):
    return str(round((tps[ac-1]-en)*100/en,2))+"%"
```

```python
def Add_trade():
    trade = []
    coin = input("enter coin")
    en = float(input("Enter Entry point"))
    tp = list(map(float, input("add takeprofit").split()))
    ac = int(input("What tp has achived"))
    trade.append(coin)
    trade.append(en)
    for i in range(6):
        if i<len(tp):
            trade.append(tp[i])
        else:
            trade.append(0)
    trade.append(ac)
    trade.append(calc_prof(en , tp , ac))
    print(len(trade))
    Trades.loc[len(Trades)] = trade
```

```python
def edit_ac(coin_name):
    print(Trades[][Trades["Coins"] == coin_name])
```
