```python
import pandas as pd
```

```python
Trades = pd.read_csv("Crypto.csv" )
```

```python
Trades
```

```python
def calc_prof(en , tps , ac):
    if ac == 0: 
        return str(0)+"%"
    return str(round((tps[ac-1]-en)*100/en,2))+"%"
```

```python
def Add_trade():
    trade = []
    coin = input("enter coin")
    en = float(input("Enter Entry point"))
    tp = list(map(float, input("add takeprofit").split()))
    ac = int(input("What tp has achived"))
    trade.append(coin)
    trade.append(en)
    for i in range(6):
        if i<len(tp):
            trade.append(tp[i])
        else:
            trade.append(0)
    trade.append(ac)
    trade.append(calc_prof(en , tp , ac))
    print(len(trade))
    Trades.loc[len(Trades)] = trade
```

```python
def edit_ac(coin_name):
    ac = int(input("Add the tp achived"))
    coin = Trades["Coins"] == coin_name
    Trades["Ac"][coin]=ac
    Trades["Profit"][Trades["Coins"] == coin_name] = calc_prof(Trades["Entry Point"][coin] , Trades[["TP1","TP2","TP3","TP4","TP5","TP6"]][coin],ac)
```

```python
Trades.drop(len(Trades)-1 , inplace=True)
```

```python
Add_trade()
```

```python
Trades.to_csv("Crypto.csv" , index=False)
```
